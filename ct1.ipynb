```python
# 1546 평균
num = int(input())
score = list(map(int,input().split()))
m=max(score)
for i in range(num):
    score[i]=score[i]/m*100
print(sum(score)/num)
```

```python
#4673 셀프 넘버
def d(n):
    ans=n
    for i in range(0,len(str(n))):
        ans += int(n/10**i)%10
        return ans
self_number=list()
for i in range(1,10001):
    self_number.append(d(i))
for i in range(1,10001):
    if i not in self_number:
        print(i)
```

```python
#2775 부녀회장이 될테야
n=int(input())
for i in range(n):
    floor=int(input()) #층
    num= int(input()) #호
    floor_0=[x for x in range(1,num+1)]
    for j in range(floor):
        for k in range(1,num):
            floor_0[k] +=floor_0[k-1]
    print(floor_0[-1])
```

```python
#2839 설탕배달
a=int(input())
cnt=0
while True:
    if a%5==0:
        a-=5
        cnt+=1
    elif a%3 ==0:
        a-=3
        cnt+=1
    else:
        a-=3
        cnt+=1
    if a==0:
        break
    if a<0:
        cnt=-1
        break
print(cnt)
        
```
